# Getting Started With Langchain

#What is langchain?
A frame-work for developing Applications[both web and stationary] usually powered by LLMs(Large Language Models).These applications will have the capability to be Aware of context and Reason(think) like a Humanoid Artificial Intelligence.

### What is a API-Key?
Application Programming Interface key, is a unique identification acts as a security credential for accessing an API (Application Programming Interface).

**Purpose**
* `Authentication` - To grant access and prevent unauthorized access
* `Tracking and control` - Track and monitor API usage based on the key used.

**To access the llms you will be needing a key, which you can get from here:**
https://platform.openai.com/api-keys

```python
# Load the required llm from the langchain_openai using the key
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(openai_api_key='.......Specify the key........')
```

```python
# Create a prompt so that the llm can better understand and interpret the query to give out the best result.
from langchain.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a proficient Technical document writer."),
    ("user","{input}")
])
```

## What is a Chain in langchain excatly?
A series of inter-connected compontnts to perform the tasks in a specific order.
* The `|` operator in LangChain syntax links these components to define the order of the execution sequence.
* Each component in the chain receives the output of the previous one as its input.
* The final output of the chain is generated by the last component.

```python
# create a simple chain combaining both llm and prompt
chain = prompt | llm
```

```python
# the output you are seeing is a AI-Chat message which is not that friendly to read
chain.invoke({"input": "how can langchain help in developing cutting-edge applications"})
```

```python
# to make the reason(output) more readable.
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt | llm | output_parser
```

```python
chain.invoke({"input": "how can langchain help in developing cutting-edge applications?"})
```
